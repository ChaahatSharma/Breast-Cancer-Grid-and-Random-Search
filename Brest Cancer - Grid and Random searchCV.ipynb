```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
bc = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
```

```python
bc.head(5)
```

```python
bc['label'] = breast_cancer.target
```

```python
bc.head(5)
```

```python
bc.shape
```

```python
bc['label'].value_counts()
```

#### 1 - Benign
#### 0 - Malignant

```python
X = bc.drop(columns = 'label', axis = 1)
Y = bc['label']
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
X
```

```python
Y
```

### Grid Search CV

```python
# Loading SVC

model = SVC()
```

```python
# Hyperparams

params = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1, 3, 6, 10]
}
```

```python
# grid search
classify = GridSearchCV(model, params, cv = 3)
```

```python
# fitting the data to our model

classify.fit(X, Y)
```

```python
classify.cv_results_
```

```python
classify.best_estimator_
```

```python
classify.best_score_
```

```python
classify.refit_time_
```

```python
results = pd.DataFrame(classify.cv_results_)
```

```python
results
```

```python
small = results[['param_C','param_kernel','mean_test_score','rank_test_score']]
```

```python
small
```

```python
sorted_df = small.sort_values(by = ['rank_test_score'])
sorted_df = sorted_df.reset_index(drop = True)
sorted_df
```

### RandomizedSearch CV

```python
# Loading SVC

model = SVC()
```

```python
# Random search
classify = RandomizedSearchCV(model, params, cv = 3)

# fitting the data to our model

classify.fit(X, Y)
```

```python
classify.cv_results_
```

```python
classify.best_estimator_
```

```python
classify.best_score_
```
